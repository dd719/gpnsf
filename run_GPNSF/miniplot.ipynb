```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
methods = ['GPNSF', 'GPNSF_nogp']
spatial_accuracy = [93.7, 87.9]
ari_scores = [0.53, 0.49]

# 设置图形大小
plt.figure(figsize=(6, 6))

# 设置柱状图的宽度和位置
x = np.arange(len(methods)) /5
width = 0.05

# 创建子图
fig, ax1 = plt.subplots(figsize=(10, 6))

# 绘制空间区域划分正确率柱状图（左轴）
bars1 = ax1.bar(x - width/2, spatial_accuracy, width, label='Acc of spatial domain identification(%)', color='#4e79a7', edgecolor='black', alpha=0.8)

# 设置左y轴标签和刻度
ax1.set_xlabel('Methods', fontsize=12, fontweight='bold')
ax1.set_ylabel('Acc(%)', fontsize=12, fontweight='bold', color='#4e79a7')
ax1.tick_params(axis='y', labelcolor='#4e79a7')
ax1.set_xticks(x)
ax1.set_xticklabels(methods, fontsize=11)
ax1.set_ylim(80, 100)  # 设置合适的y轴范围

# 在柱状图上添加数值标签
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=10, fontweight='bold', color='#4e79a7')

# 创建第二个y轴（右轴）用于ARI分数
ax2 = ax1.twinx()
bars2 = ax2.bar(x + width/2, ari_scores, width, label='ARI', color='#e15759', edgecolor='black', alpha=0.8)

# 设置右y轴标签和刻度
ax2.set_ylabel('ARI', fontsize=12, fontweight='bold', color='#e15759')
ax2.tick_params(axis='y', labelcolor='#e15759')
ax2.set_ylim(0, 0.6)  # 设置合适的y轴范围

# 在柱状图上添加数值标签
for bar in bars2:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold', color='#e15759')

# 添加图例
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=10)

# # 添加标题
# plt.title('GPNSF与GPNSF_nogp方法性能比较', fontsize=14, fontweight='bold', pad=20)

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()

# 或者保存图形
# plt.savefig('method_comparison.png', dpi=300, bbox_inches='tight')
# plt.close()
```
